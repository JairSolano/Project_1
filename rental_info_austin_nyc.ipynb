```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read in CSV
rental_df = pd.read_csv("../Project-1---Citywide-Payroll/Rental Info/Metro_zori_uc_sfrcondomfr_sm_month.csv")
rental_df
```

```python
rental_df.columns
```

```python
#read in CSV
rental_zip = pd.read_csv("../Project-1---Citywide-Payroll/Rental Info/Zip_zori_uc_sfrcondomfr_sm_month.csv")
rental_zip
```

```python
#start filtering data to get NYC data only
ny_rental = rental_zip.loc[rental_zip['StateName'] == 'NY']
ny_rental
```

```python
#nyc_rental = ny_rental.loc[ny_rental['CountyName'] == 'Bronx County' or 'Kings County' or 'Manhattan County' or 'Queens County' or 'Staten Island County']
#filtering continued resulting in nyc 5 borough counties.
nyc_rental = ny_rental.loc[ny_rental['CountyName'].isin(['Bronx County', 'Kings County', 'New York County', 'Queens County', 'Richmond County'])]
nyc_rental.columns

```

```python
#rename columns for cleaner appearance
#using 09-30 for date of rent for all years
nyc_rental_yrs = nyc_rental[['RegionName',
       'State', 'City','CountyName', '2023-09-30','2022-09-30','2021-09-30', '2020-09-30', '2019-09-30','2018-09-30','2017-09-30','2016-09-30','2015-09-30']]
nyc_rental_yrs['CountyName'].unique()
nyc_rental_yrs.rename(columns={'2023-09-30':'2023','2022-09-30':'2022','2021-09-30':'2021', '2020-09-30':'2020', '2019-09-30':'2019','2018-09-30':'2018','2017-09-30':'2017','2016-09-30':'2016','2015-09-30':'2015'}, inplace=True)
nyc_rental_yrs.columns
```

```python
#cleanig up data
#drop NA 
#sorted by CountyName
nyc_rental_yrs_sorted = nyc_rental_yrs.sort_values(by='CountyName')
nyc_rental_yrs_sorted = nyc_rental_yrs_sorted.dropna()
nyc_rental_yrs_sorted
```

```python
#Finding the data for Austin TX as a comparative city.
austin_rental = rental_zip.loc[rental_zip['City'] == 'Austin']
austin_rental = austin_rental.loc[austin_rental['State'] == 'TX']
austin_rental = austin_rental.dropna()
```

```python
#Renamed columns to match NYC info and cleaner appearance.
austin_rental_yrs = austin_rental[['RegionName',
       'State', 'City','CountyName', '2023-09-30','2022-09-30','2021-09-30', '2020-09-30', '2019-09-30','2018-09-30','2017-09-30','2016-09-30','2015-09-30']]
austin_rental_yrs.rename(columns={'2023-09-30':'2023','2022-09-30':'2022','2021-09-30':'2021', '2020-09-30':'2020', '2019-09-30':'2019','2018-09-30':'2018','2017-09-30':'2017','2016-09-30':'2016','2015-09-30':'2015'}, inplace=True)
austin_rental_yrs.columns
```

```python
rent_by_borough = nyc_rental_yrs_sorted.set_index(['CountyName'])
rent_by_borough['2018'].max()

```

```python
rent_by_county = austin_rental_yrs.groupby(['CountyName'])
rent_by_county['2023'].mean()
```

```python
borough_df = pd.DataFrame(rent_by_borough)
borough_df.head()
```

```python
rental_pivot = pd.pivot_table(nyc_rental_yrs_sorted, columns='CountyName', values=['2023','2022','2021','2020','2019','2018','2017','2016','2015'])

rental_pivot
```

```python
plt.plot(rental_pivot)
plt.legend(loc='best')

plt.savefig('nyc_rent_avg.jpeg')
plt.show()
```

```python
avg_rent_by_borough = rent_by_borough[[str(year) for year in range(2015,2024)]].mean()
avg_rent_by_borough
```

```python
years = avg_rent_by_borough.keys()
years
```

```python
plt.plot(years, avg_rent_by_borough, color='blue', label='NYC')
plt.plot(austin_pivot, color='red', label='Austin')
plt.title('Average Monthly Rent')
plt.xlabel('Years')
plt.ylabel('Rent in $')
plt.legend(loc='best')
plt.savefig('avg_rents.jpeg')
#plt.plot(years, avg_rent_by_borough.loc['Bronx County',['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']], color='green',label='Bronx')
#plt.plot(years, avg_rent_by_borough.loc['Queens County',['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']], color='red',label='Queens')
#plt.plot(years, avg_rent_by_borough.loc['Richmond County',['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']], color='lightblue',label='Staten Island')
#plt.plot(years, avg_rent_by_borough.loc['Kings County',['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']], color='pink',label='Brooklyn')
#plt.plot(years, avg_rent_by_borough.loc['New York County',['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']], color='purple',label='Manhattan')
plt.show()
```
